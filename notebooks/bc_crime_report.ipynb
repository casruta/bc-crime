## Executive Summary

This analysis examines four core questions about crime in British Columbia:

1. **Is crime rising?** BC's Crime Severity Index peaked in 1998 (166.9) and fell 46% to a trough in 2014 (90.2). Since then it has plateaued around 93–107, with a notable 7.4% drop in the latest year (2024). BC's overall crime rate remains 21% above the national average.

2. **What kinds of crime are changing?** Property crime dominates (51% of total rate) but is declining — theft from vehicles fell by 772 per 100,000 over five years. Meanwhile, child exploitation offences (+82/100k) and shoplifting (+79/100k) are rising sharply. Violent crime's share is growing even as its absolute rate stabilizes.

3. **What does policing cost?** Real policing expenditure in BC grew 145% from 1986–2009, even as the CSI was falling. In 2009, BC spent $253 per capita on policing — 31% below the national average of $366. Post-2009 provincial expenditure data is unavailable from StatCan.

4. **Where is crime concentrated?** Vancouver (48,812 offences) and Surrey (41,275) account for the bulk of BC's crime volume. But per-capita rates tell a different story: interior CMAs like Chilliwack (11,352/100k) and Kamloops (10,546/100k) have twice the crime rate of Vancouver (5,438/100k) and Victoria (5,283/100k).

## Methodology Notes

### Data Sources

| Source | Table | Description | Coverage |
|--------|-------|-------------|----------|
| Statistics Canada | 35-10-0063-01 | Crime Severity Index by police service, BC | 1998–2024 |
| Statistics Canada | 35-10-0177-01 | Incident-based crime stats by province/CMA | 1998–2024 |
| Statistics Canada | 35-10-0076-01 | Police personnel and expenditures | 1986–2023 |
| Statistics Canada | 35-10-0059-01 | Police services expenditures (Canada) | 2018–2023 |
| Statistics Canada | 18-10-0005-01 | Consumer Price Index, annual | 1914–2024 |
| BC Gov | Appendix F | Crime Statistics in BC, 2023 | 2022–2023 |
| BC Gov | Appendix G | BC Crime Trends | 2014–2023 |
| BC Gov | Appendix I | BC Jurisdiction Crime Trends | 2014–2023 |

### Key Metrics

- **Crime Severity Index (CSI):** Developed by StatCan, assigns a weight to each offence based on average sentences. Higher = more serious crime mix. Base year: 2006.
- **Crime rate:** Criminal Code offences per 100,000 population, excluding traffic violations.
- **Constant dollars:** Nominal expenditure figures adjusted using the All-Items CPI (2002=100 base), rebased to 2020 dollars.

### Caveats and Limitations

1. **Police-reported data only.** These statistics reflect crimes reported to and recorded by police. Unreported crime (the "dark figure") is not captured. Victimization surveys suggest only about 30% of crimes are reported.

2. **Reporting changes.** Changes in police recording practices, public reporting behaviour, and legislative amendments can affect statistics independently of actual crime trends. For example, the 2019 reclassification of some assault charges affected comparability.

3. **Population estimates.** Rates per 100,000 use mid-year population estimates, which are revised after each census. Inter-censal estimates may be less accurate for fast-growing communities.

4. **COVID-19 anomaly.** The 2020 data point reflects unusual conditions (lockdowns, reduced mobility, closed businesses) that temporarily suppressed some crime categories while potentially increasing others (domestic violence).

5. **Expenditure data gap.** Provincial-level policing expenditure data from StatCan ends at 2009. Post-2009 comparisons use Canada-wide aggregates only.

6. **Jurisdiction boundaries.** Police jurisdiction boundaries do not always align with municipal boundaries. Some RCMP detachments cover multiple communities. Comparing jurisdictions by raw count without population data can be misleading.
